```python
import gurobipy as gurobi
```

```python
random_city_codes = random.sample(range(1,82),15)
```

```python
random_cities = []
for i in range(len(random_city_codes)):
    random_cities.append(province_data[province_data["İL NO"] == random_city_codes[i]]["İL ADI"]\
                         [random_city_codes[i]-1])
random_cities
```

```python
random_distance = {(c1, c2): province_data[province_data["İL ADI"] == c1][c2]\
                   [province_data[province_data["İL ADI"] == c1]["İL NO"].index[0]]\
                       for c1, c2 in product(random_cities, repeat= 2)}
random_distance
```

```python
#Model Construction
from gurobipy import GRB

m = gurobi.Model()

vars = m.addVars(random_distance.keys(),obj = random_distance, vtype=GRB.BINARY, name='x')

for i, j in vars.keys():
    vars[j, i] = vars[i, j]

cons1 = m.addConstrs(vars.sum(c, '*') == 2 for c in random_cities)
```

```python
# Callback - use lazy constraints to eliminate sub-tours

def subtourelim(model, where):
    if where == GRB.Callback.MIPSOL:
        # make a list of edges selected in the solution
        vals = model.cbGetSolution(model._vars)
        selected = gurobi.tuplelist((i, j) for i, j in model._vars.keys()
                             if vals[i, j] > 0.5)
        # find the shortest cycle in the selected edge list
        tour = subtour(selected)
        if len(tour) < len(cities):
            # add subtour elimination constr. for every pair of cities in subtour
            model.cbLazy(gurobi.quicksum(model._vars[i, j] for i, j in combinations(tour, 2))
                         <= len(tour)-1)

# Given a tuplelist of edges, find the shortest subtour

def subtour(edges):
    unvisited = cities[:]
    cycle = cities[:] # Dummy - guaranteed to be replaced
    while unvisited:  # true if list is non-empty
        thiscycle = []
        neighbors = unvisited
        while neighbors:
            current = neighbors[0]
            thiscycle.append(current)
            unvisited.remove(current)
            neighbors = [j for i, j in edges.select(current, '*')
                         if j in unvisited]
        if len(thiscycle) <= len(cycle):
            cycle = thiscycle # New shortest subtour
    return cycle
```

```python
m._vars = vars
m.Params.lazyConstraints = 1
m.optimize(subtourelim)
```

```python
# Retrieve solution

vals = m.getAttr('x', vars)
selected = gurobi.tuplelist((i, j) for i, j in vals.keys() if vals[i, j] > 0.5)

tour = subtour(selected)
#assert len(tour) == len(cities)

for i in random_cities:
    for j in random_cities:
        if vals[(i,j)] == 1:
            print("(" + i + "," + j + ")" + "=" +" " + "1")
```
